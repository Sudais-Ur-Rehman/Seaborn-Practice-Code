Create a count plot to show frequency of categories.

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

df = sns.load_dataset('tips')
df
```

```python
sns.countplot(x='time',
              hue='sex',
              data=df,
              palette='pastel')
plt.title('Number of Customers by Time and Gender')
plt.xlabel('Time of Day')
plt.ylabel('Count')
plt.show()
```
