Create a histogram using seaborn.

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

df = sns.load_dataset('penguins')
df = df.dropna(subset=['flipper_length_mm'])
```

```python
sns.histplot(x='flipper_length_mm',
             bins=10,
             hue='species',
             data=df,
            multiple='stack',  # You can also try 'dodge' or 'layer'
            palette='Set2')
plt.title('Flipper Length Distribution by Species')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```
