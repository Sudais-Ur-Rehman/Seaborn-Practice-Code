Create a line plot with confidence intervals.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = sns.load_dataset('fmri')  # Built-in dataset with time-series data

# Create line plot with confidence intervals
sns.lineplot(
    x='timepoint',
    y='signal',
    hue='event',
    data=df,
    palette='Set2',
    errorbar='ci'  # Default: 95% confidence interval
)

# Customize plot
plt.title('FMRI Signal Over Time with Confidence Intervals')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.grid(True)
plt.show()
```
