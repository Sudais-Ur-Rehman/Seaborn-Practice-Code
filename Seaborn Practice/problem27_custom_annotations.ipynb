Add custom text and annotations to seaborn plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

# Load dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/monthly-sunspots.csv"
df = pd.read_csv(url)
df["Month"] = pd.to_datetime(df["Month"])

# Find the peak value
max_idx = df["Sunspots"].idxmax()
max_month = df.loc[max_idx, "Month"]
max_value = df.loc[max_idx, "Sunspots"]

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(
    x="Month",
    y="Sunspots",
    data=df,
    color="darkblue",
    marker="o",
    label="Monthly Sunspots"
)

# Highlight the max point
plt.scatter(max_month, max_value, color="red", s=100, zorder=5)

# Add annotation
plt.annotate(
    f"Peak: {max_value} in {max_month.strftime('%Y-%m')}",
    xy=(max_month, max_value),         # point to annotate
    xytext=(max_month, max_value + 50),# text location
    arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8),
    fontsize=12,
    color="darkred",
    fontweight="bold"
)

# Labels and title
plt.title("Monthly Sunspots with Peak Highlight", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Sunspots", fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
