Use hue and dodge to separate data in barplot or boxplot.

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

df = sns.load_dataset('tips')
df
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Barplot
sns.barplot(x="time",
            y="total_bill",
            hue="sex",
            dodge=True,
            data=df,
            palette="pastel",
            ax=axes[0])
axes[0].set_xlabel("Time")
axes[0].set_ylabel("Total Bill")
axes[0].set_title("Total Bill Over Time - Barplot")

# Boxplot
sns.boxplot(x="time",
            y="total_bill",
            hue="sex",
            dodge=True,
            data=df,
            palette="pastel",
            ax=axes[1])
axes[1].set_xlabel("Time")
axes[1].set_ylabel("Total Bill")
axes[1].set_title("Total Bill Over Time - Boxplot")

plt.tight_layout()
plt.show()
```
