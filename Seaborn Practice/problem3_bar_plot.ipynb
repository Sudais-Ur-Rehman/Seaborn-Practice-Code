Create a bar plot using seaborn.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 

df = sns.load_dataset('tips')
df
```

```python
sns.barplot(x='time',
            y='total_bill',
            hue='sex',
            data=df,
            errorbar='sd',
            palette='pastel')
plt.title('Average Total Bill by Time and Gender')
plt.xlabel('Time of Day')
plt.ylabel('Average Total Bill')
plt.grid(axis='y')
plt.show()
```
