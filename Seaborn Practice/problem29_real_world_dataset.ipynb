Analyze and visualize a real-world dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load a real-world dataset
df = sns.load_dataset("penguins")

# Basic info about the dataset
print("Dataset Overview:")
print(df.head(), "\n")
print(df.info(), "\n")
print(df.describe())

# Handle missing values (drop rows with NaN for simplicity)
df = df.dropna()

# -------------------------------
# 1. Count plot - species distribution
plt.figure(figsize=(6,4))
sns.countplot(x="species", data=df, palette="pastel")
plt.title("Penguin Species Count")
plt.show()

# -------------------------------
# 2. Scatter plot - bill length vs bill depth
plt.figure(figsize=(6,4))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", hue="species", data=df, palette="muted")
plt.title("Bill Length vs Bill Depth by Species")
plt.show()

# -------------------------------
# 3. Box plot - flipper length distribution by island
plt.figure(figsize=(6,4))
sns.boxplot(x="island", y="flipper_length_mm", data=df, palette="Set2")
plt.title("Flipper Length Distribution by Island")
plt.show()

# -------------------------------
# 4. Pair plot - relationship between multiple features
sns.pairplot(df, hue="species", palette="Set1")
plt.show()
```
