Create a pairplot to explore relationships in a dataset.

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

df = sns.load_dataset("iris")
df = df.rename(columns={
    'sepal_length': 'Sepal Length (cm)',
    'sepal_width': 'Sepal Width (cm)',
    'petal_length': 'Petal Length (cm)',
    'petal_width': 'Petal Width (cm)',
    'species': 'Species'
})
```

```python
g = sns.pairplot(data=df, hue='Species')

# Move legend outside
g._legend.set_bbox_to_anchor((1.05, 0.5))  # x=1.05 means just outside the plot
g._legend.set_title("Species")
plt.suptitle("Iris Flower Measurements - Pairwise Relationships", y=1.02)
plt.tight_layout()
plt.show()
```
