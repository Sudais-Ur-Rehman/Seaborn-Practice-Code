Load a CSV file and create seaborn visualizations.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset from GitHub
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
df = pd.read_csv(url)

# Drop rows with missing values for simplicity
df = df.dropna()

print(df.head())

# Set style
sns.set_theme(style="whitegrid")

# 1. Bar plot with error bars
plt.figure(figsize=(8, 5))
sns.barplot(
    x="species",
    y="body_mass_g",
    data=df,
    errorbar=("ci", 95),
    palette="pastel"
)
plt.title("Body Mass by Penguin Species (95% CI)")
plt.show()

# 2. Grouped bar plot with hue
plt.figure(figsize=(8, 5))
sns.barplot(
    x="species",
    y="body_mass_g",
    hue="sex",
    data=df,
    errorbar=("ci", 95),
    palette="muted"
)
plt.title("Body Mass by Species & Sex")
plt.show()

# 3. Scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species",
    style="sex",
    data=df,
    palette="deep"
)
plt.title("Flipper Length vs Body Mass")
plt.show()
```
