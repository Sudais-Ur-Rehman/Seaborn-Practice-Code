Visualize time series data using seaborn.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/monthly-sunspots.csv"
df = pd.read_csv(url)
df["Month"] = pd.to_datetime(df["Month"])
df.head()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(
    x="Month",
    y="Sunspots",
    data=df,
    color="crimson",           # custom color
    marker="o",                # circle markers
    markersize=4,              # marker size
    label="Monthly Sunspots",  # legend label
    linewidth=1.5)
# Labels and title
plt.title("Monthly Sunspots Over Time", fontsize=16, fontweight="bold")
plt.xlabel("Date", fontsize=12)
plt.ylabel("Number of Sunspots", fontsize=12)

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Show legend
plt.legend()

# Show plot
plt.show()
```
