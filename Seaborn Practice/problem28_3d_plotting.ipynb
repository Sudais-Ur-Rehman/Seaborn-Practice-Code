Use seaborn with matplotlib for basic 3D visualizations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D  # Enables 3D plotting in Matplotlib

# Apply seaborn theme
sns.set(style="whitegrid")

# Create some random 3D data
np.random.seed(42)
x = np.random.rand(50) * 10
y = np.random.rand(50) * 10
z = np.sin(x) + np.cos(y)  # a function for variation

# Create 3D figure
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")

# Scatter plot
scatter = ax.scatter(
    x, y, z,
    c=z,                # color based on z values
    cmap="viridis",     # colormap
    s=50,               # point size
    edgecolor="k"       # black border around points
)

# Labels
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
ax.set_title("Basic 3D Scatter Plot with Seaborn Style")

# Add color bar
cbar = fig.colorbar(scatter, ax=ax, shrink=0.5, aspect=10, pad=0.1)
cbar.set_label("Z Value", rotation=270, labelpad=15)

plt.show()
```
