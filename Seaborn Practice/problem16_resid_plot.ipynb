Create a residual plot using seaborn.

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

df = sns.load_dataset('mpg')
df 
```

```python
sns.residplot(x='horsepower',
              y='mpg',
              data=df,
              color="blue",
              scatter_kws={"alpha":0.6})
plt.show()
```
