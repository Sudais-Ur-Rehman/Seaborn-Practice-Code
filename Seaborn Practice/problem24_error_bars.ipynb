Add error bars to bar plots using `ci` or custom errors.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load example dataset
df = sns.load_dataset('tips')

# Bar plot with modern confidence intervals
sns.barplot(
    x="day",
    y="total_bill",
    data=df,
    errorbar=('ci', 95),   # Replaces ci=95
    palette="pastel",
    legend=False
)
plt.title("Bar Plot with 95% CI")
plt.show()

# Custom error bars using Matplotlib (mean Â± std)
means = df.groupby("day")["total_bill"].mean()
errors = df.groupby("day")["total_bill"].std()

plt.bar(
    x=means.index,
    height=means.values,
    yerr=errors.values,
    capsize=5,
    color=sns.color_palette("muted")
)
plt.title("Custom Error Bars (Std Dev)")
plt.ylabel("Total Bill")
plt.show()
```
