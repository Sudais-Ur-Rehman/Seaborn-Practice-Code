Use catplot to visualize categorical data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = sns.load_dataset('tips')

# Create a categorical plot (bar plot)
sns.catplot(
    x='day',
    y='total_bill',
    hue='sex',
    kind='bar',       # You can also try 'violin', 'swarm', 'box', 'strip'
    data=df,
    palette='pastel',
    errorbar='sd'     # Show standard deviation
)

# Customize
plt.title("Average Total Bill by Day and Gender")
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill ($)")
plt.show()
```
