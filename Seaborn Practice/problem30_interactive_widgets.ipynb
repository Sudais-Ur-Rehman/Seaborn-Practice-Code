Use ipywidgets with seaborn for interactive plots.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# Load dataset
df = sns.load_dataset("penguins").dropna()

# Function to create interactive plot
def interactive_plot(plot_type, x_axis, y_axis, hue, palette):
    plt.figure(figsize=(7,5))
    
    if plot_type == "Scatter":
    
        sns.scatterplot(x=x_axis, y=y_axis, hue=hue, data=df, palette=palette)
    elif plot_type == "Box":
        sns.boxplot(x=x_axis, y=y_axis, hue=hue, data=df, palette=palette)
    
    elif plot_type == "Violin":
        sns.violinplot(x=x_axis, y=y_axis, hue=hue, data=df, palette=palette)
    elif plot_type == "Bar":
        sns.barplot(x=x_axis, y=y_axis, hue=hue, data=df, palette=palette)
    
    plt.title(f"{plot_type} Plot of {y_axis} vs {x_axis}")
    plt.show()

# Widgets for interactivity
plot_type_widget = widgets.Dropdown(options=["Scatter", "Box", "Violin", "Bar"], value="Scatter", description="Plot Type:")
x_axis_widget = widgets.Dropdown(options=df.columns, value="bill_length_mm", description="X Axis:")
y_axis_widget = widgets.Dropdown(options=df.columns, value="bill_depth_mm", description="Y Axis:")
hue_widget = widgets.Dropdown(options=[None] + list(df.columns), value="species", description="Hue:")
palette_widget = widgets.Dropdown(options=["Set1", "Set2", "muted", "pastel", "coolwarm"], value="Set1", description="Palette:")

# Create interactive output
interact(interactive_plot,
         plot_type=plot_type_widget,
         x_axis=x_axis_widget,
         y_axis=y_axis_widget,
         hue=hue_widget,
         palette=palette_widget)
        
```
