Create a box plot to show distribution.

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

df = sns.load_dataset("tips")
```

```python
sns.boxplot(x='time',y='total_bill',data=df,hue='sex',palette='pastel')
plt.title("Total Bill Distribution by Time and Gender")
plt.xlabel("Time of Day")
plt.ylabel("Total Bill ($)")
plt.grid(axis='y')      
plt.show()
```
