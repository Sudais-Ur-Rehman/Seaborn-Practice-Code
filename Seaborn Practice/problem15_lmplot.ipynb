Create a linear model plot (lmplot).


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create a linear model plot
sns.lmplot(
    x="total_bill",
    y="tip",
    data=tips,
    hue="sex",        # Different colors for each sex
    markers=["o", "s"], # Marker shapes
    palette="Set1",     # Color palette
    height=6,           # Figure height
    aspect=1.2          # Width-to-height ratio
)

plt.title("Linear Model Plot of Total Bill vs Tip")
plt.show()
```
