Create a swarm plot.

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

df = sns.load_dataset('penguins')
df
```

```python
sns.swarmplot(
    x="species",
    y="bill_length_mm",
    data=df,
    palette="pastel"
)

# Customize plot
plt.title("Bill Length Distribution by Penguin Species")
plt.xlabel("Species")
plt.ylabel("Bill Length (mm)")
plt.show()

```
